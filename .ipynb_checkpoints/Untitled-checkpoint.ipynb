```python
import pandas as pd
df = pd.read_csv("C:/Users/User/OneDrive/Desktop/Projects/Sales/customer_shopping_behavior.csv")
df.head()
```

```python
df.describe(include ='all')
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount(usd)'})
```

```python
df.columns
```

```python
labels=['Young Adult' , 'Adult','Middle Aged' , 'Senior Citizen']
df['age_group'] = pd.qcut(df['age'] , q=4 ,labels =labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchases_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchases_frequency_days','frequency_of_purchases']]
```
